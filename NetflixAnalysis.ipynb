### All weeks countries

```python
import pandas as pd
import pycountry_convert as pc
```

```python
allweekscountries = pd.read_excel("Jun2024/all-weeks-countries.xlsx",sheet_name="Top 10")
```

#### add continent

```python
def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name
```

```python
allweekscountries["Continent"] = allweekscountries["country_name"].apply(lambda x : country_to_continent(x))
```

```python
allweekscountries["Continent"].unique().tolist()
```

```python
allweekscountries=allweekscountries.fillna("")
```

#### split the seasons

```python
def splittheseason(x):
    season = ""
    if len(x.split("Season")) > 1:
        season = x.split("Season")[1].strip()
    elif len(x.split("Part")) > 1:
        season = x.split("Part")[1].strip()
    elif len(x.split("Limited Series")) > 1:
        season = x.split('Limited Series')[1].strip()
        
    return season
```

```python
allweekscountries["Season"] = allweekscountries["season_title"].apply(lambda x : splittheseason(x))
```

```python
# allweekscountries["Season"].unique().tolist()
```

```python
allweekscountries.to_csv("Jun2024/all-weeks-countries-edited.csv",index=False)
```

#### country of origin

```python
initialcountryoforiginandshowtype = pd.read_csv("initialcountryoforiginandshowtype.csv")
combinedbycountryandshow = allweekscountries.merge(initialcountryoforiginandshowtype, left_on=["show_title"], right_on=["show_title"], how="left")
# combinedbycountryandshowv1 = combinedbycountryandshow[["show_title", "show_type", "country_of_origin"]]
combinedbycountryandshowv1=combinedbycountryandshow.fillna("")
combinedbycountryandshowv1=combinedbycountryandshowv1.drop_duplicates()
# combinedbycountryandshowv1.to_csv("combinedbycountryandshowv1.csv",index=False)
```
